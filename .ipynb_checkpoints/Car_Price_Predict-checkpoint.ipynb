```python
df = data[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Fuel_Type',
       'Seller_Type', 'Transmission', 'Owner']]
```

```python
df.head()
```

```python
df['Current_Year']=2022
```

```python
df.head()
```

```python
df['No_Year']=df['Current_Year']-df['Year']
```

```python
df.head()
```

```python
df.drop(['Year','Current_Year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
# Finding Corelation

df.corr()
```

```python
snb.pairplot(df)
```

```python
corrmat = df.corr()
top_corelation_features = corrmat.index
mpl.figure(figsize=(20,20))

#plot heat map
g = snb.heatmap(df[top_corelation_features].corr(),annot=True,cmap='RdYlGn')
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
X.columns
```

```python
y.head()
```

```python
y.shape
```

```python
X.shape
```

### Features Importance

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
```

```python
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

### Plot the most important features

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(5).plot(kind='barh')
mpl.show()
```
